```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
print("ðŸ”¹ First 5 rows of the dataset:")
print(df.head())

print("\nðŸ”¹ Dataset Info:")
print(df.info())


print("\nðŸ”¹ Statistical Summary:")
print(df.describe())


print("\nðŸ”¹ Missing Values:")
print(df.isnull().sum())

print("\nðŸ”¹ Fertilizer Name Distribution:")
print(df['Fertilizer Name'].value_counts())
```

```python
lt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Fertilizer Name', order=df['Fertilizer Name'].value_counts().index)
plt.title('Fertilizer Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Heatmap for correlations (for numeric features)
numeric_features = ['Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Phosphorous', 'Potassium']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Box plots for numerical features vs Fertilizer
for feature in numeric_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Fertilizer Name', y=feature, data=df)
    plt.title(f'{feature} vs Fertilizer Name')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Categorical analysis: Soil Type
print("\nðŸ”¹ Soil Type Distribution:")
print(df['Soil Type'].value_counts())

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Soil Type', order=df['Soil Type'].value_counts().index)
plt.title('Soil Type Distribution')
plt.show()

# Categorical analysis: Crop Type
print("\nðŸ”¹ Crop Type Distribution:")
print(df['Crop Type'].value_counts())

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Crop Type', order=df['Crop Type'].value_counts().index)
plt.title('Crop Type Distribution')
plt.xticks(rotation=45)
plt.show()

```
